# Importação Biblioteca

* numpy (np): É usada para operações matemáticas e para gerar uma sequência de valores.
* matplotlib.pyplot (plt): É usada para criar gráficos e visualizações.

```python
import numpy as np
import matplotlib.pyplot as plt
```

# Definição da função Gaussiana
Esta função calcula o valor da função de densidade de probabilidade Gaussiana para uma determinada entrada
* x: Valor ou vetor de valores onde a função será avaliada.
* mu: Média da distribuição.
* sigma: Desvio padrão da distribuição.

```python
# Definindo a função de densidade de probabilidade Gaussiana
def gaussian(x, mu, sigma):
    return (1 / (np.sqrt(2 * np.pi * sigma**2))) * np.exp(-((x - mu)**2) / (2 * sigma**2))
```

# Definição dos parâmetros da distribuição

```python
# Parâmetros da distribuição
mu = 0  # Média
sigma = 5  # Desvio padrão
n = 100000
```

# Geração dos Pontos de x

* np.linspace(-5, 5, n): Gera uma sequência de n
n pontos igualmente espaçados entre -5 e 5.
* Esses pontos serão usados como os valores
x nos quais a função Gaussiana será avaliada.

```python
# Gerando n pontos da função gaussiana usando um loop for
x_values = np.linspace(-5, 5, n)  # Gera n pontos entre -5 e 5
y_values = []
```

# Cálculo dos valores y (a função Gaussiana)


```python
# Gerando a curva
y_values = gaussian(x_values, mu, sigma)
```

# Plotando a curva Gaussiana

```python
# Plotando a função gaussiana
plt.plot(x_values, y_values, label=f'Gaussian: $\mu={mu}$, $\sigma={sigma}$')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Função Gaussiana (Distribuição Normal)')
plt.legend()
plt.grid(True)
plt.show()
```
