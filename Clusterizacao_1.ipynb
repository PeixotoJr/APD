# Importação de bibliotecas


* numpy (como np): usada para criar e manipular arrays e números aleatórios.
* matplotlib.pyplot (como plt): usada para gerar gráficos.
* KMeans da biblioteca sklearn.cluster: algoritmo de clusterização K-Means.



```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

# Definição de semente aleatória
A semente (seed) é definida para garantir a reprodutibilidade dos números aleatórios gerados. Usar a mesma semente faz com que os resultados sejam consistentes em diferentes execuções.

```python
seed = 42
np.random.seed(seed)
```

# Geração dos dados (três clusters)
Aqui, são gerados três conjuntos de pontos aleatórios que seguem uma distribuição normal (ou seja, eles têm valores aleatórios baseados em uma curva gaussiana) com 100 pontos e duas dimensões cada:

* np.random.randn(100, 2) gera um array de 100 pontos com duas características (2 colunas), onde os valores são distribuídos normalmente com média 0 e desvio padrão 1.
* np.array([0, 0]) adiciona um deslocamento ao centro dos pontos, que neste caso é [0, 0].
* np.vstack((cluster_1, cluster_2, cluster_3)) junta verticalmente (empilha) os três clusters gerados em um único array X com 300 pontos no total.

```python
cluster_1 = np.random.randn(100, 2) + np.array([0, 0])
cluster_2 = np.random.randn(100, 2) + np.array([0, 0])
cluster_3 = np.random.randn(100, 2) + np.array([0, 0])
X = np.vstack((cluster_1, cluster_2, cluster_3))
```

# Visualização dos pontos gerados

```python
# Visualização
plt.figure(figsize=(10, 6))
plt.scatter(X[:, 0], X[:, 1])
#plt.scatter(centers[:, 0], centers[:, 1], c='red', marker='x', s=100, label='Centróides')
plt.title('Pontos Gerados')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()
```

# Aplicação do algoritmo K-Means

O algoritmo K-Means é aplicado aos dados:

* KMeans(n_clusters=3) inicializa o algoritmo K-Means, definindo o número de clusters como 3. O parâmetro random_state=seed garante a reprodutibilidade.
* kmeans.fit(X) executa o processo de clusterização nos dados X, ajustando os clusters e calculando as centróides.

```python
# Aplicando o KMeans (aqui você pode alterar o número de clusters)
kmeans = KMeans(n_clusters=5, random_state=seed)
kmeans.fit(X)
```

# Extração das centróides

```python
# Centróides
centers = kmeans.cluster_centers_
```

# Visualização dos clusters e centróides

```python
# Visualização
plt.figure(figsize=(10, 6))
plt.scatter(X[:, 0], X[:, 1], c=kmeans.labels_, cmap='viridis', marker='o', label='Pontos')
plt.scatter(centers[:, 0], centers[:, 1], c='red', marker='x', s=100, label='Centróides')
plt.title('Clusters com KMeans')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()
```
